## Restrictions on Dictionary Keys
- Almost any type of value can be used as a dictionary key in Python. Below example, where integer, float, and Boolean objects are used as keys

````python
d1={42: 'aaa', 2.78: 'bbb', True: 'ccc'}
````

````python
d1
````

- You can even use built-in objects like types and functions

````python
d2={int: 1, float: 2, bool: 3}
````

````python
d2[int]
````

````python
int(1.8)
````

- However, there are a couple restrictions that dictionary keys must abide by.

  - First, a given key can appear in a dictionary only once. Duplicate keys are not allowed. A dictionary maps each key to a corresponding value, so it doesn’t make sense to map a particular key more than once.

* Restrictions on Dictionary Values
  - By contrast, there are no restrictions on dictionary values. Literally none at all. A dictionary value can be any type of object Python supports, including mutable types like lists and dictionaries, and user-defined objects, which you will learn about in upcoming tutorials.

## Dictionary Methods

- As with strings and lists, there are several built-in methods that can be invoked on dictionaries. In fact, in some cases, the list and dictionary methods share the same name. (In the discussion on object-oriented programming, you will see that it is perfectly acceptable for different types to have methods with the same name.)

- items() -- method returns a view object. The view object contains the key-value pairs of the dictionary, as tuples in a list. The view object will reflect any changes done to the dictionary.

````python
dict1 = {'k1': 100,'k2':500, "k10":310,  'k3': 200,"k5":50}
dict1
````

````python
dict1.items() # when we need access to the key and value simultaneously
````

- keys() -- method returns a view object. The view object contains the keys of the dictionary, as a list.
- values() -- method returns a view object. The view object contains the values of the dictionary, as a list.

````python
dict1 = {'k1': 100,'k2':500, "k10":310,  'k3': 200,"k5":50}
dict1
````

````python
dict1.items() # Iterable - is something that you can use in a loop
````

````python
dict1.keys()
````

````python
dict1.values()
````

- fromkeys() - method is used to create a new dictionary with specified keys and a common value for all keys.

  - Key Points:
    - The method is useful for initializing dictionaries with default values.
    - The keys are shared among the same value reference for mutable types.



````python
keys = ["key1", "key2", "key3"]
new_dict = dict.fromkeys(keys)
print(new_dict)
````

````python
keys = ["name", "age", "gender"]
default_value = "Not Specified"
new_dict = dict.fromkeys(keys, default_value)
print(new_dict)
````

- setdefault() - method returns the value of a specified key in a dictionary. If the key does not exist, it inserts the key with a specified default value.


````python
my_dict = {"name": "Alice", "age": 25}
result = my_dict.setdefault("name", "Default")
print(result)
````

````python
print(my_dict)
````

````python
my_dict = {"name": "Alice", "age": 25}
result = my_dict.setdefault("gender", "Female")
print(result)
````

````python
print(my_dict)
````

````python
my_dict = {"name": "Alice", "age": 25}
result = my_dict.setdefault("address")
print(result)
````

````python
print(my_dict)
````

## Nested Dictionary
- In Python, a nested dictionary is a dictionary inside a dictionary. It's a collection of dictionaries into one single dictionary.

- Here, the nested_dict is a nested dictionary with the dictionary dictA and dictB. They are two dictionary each having own key and value.

### Access elements of a Nested Dictionary
- To access element of a nested dictionary, we use indexing [] syntax in Python.

````python
people = {1: {'name': 'John', 'age': '27', 'sex': 'Male'},
          2: {'name': 'Marie', 'age': '22', 'sex': 'Female'}}
````

````python
people[2]['age']
````

### Add element to a Nested Dictionary

````python
# change the second name - from Marie - Maria
# add another name to the dictionary - assign a dynamic index
# add new key value pair to the latest record added above - which included phone number
````

````python
people[2]["name"]="Maria"
````

````python
people[max(people.keys())+1] = {'name': 'Rakib', 'age': '25', 'sex': 'Male'}
````

````python
people.update({ max(people.keys())+1 : {'name': 'Rakib', 'age': '25', 'sex': 'Male'}})
````

````python
people[max(people.keys())]['Phone'] = 9843957352
````

````python
people
````

````python
# Add another dictionary to the nested dictionary
people = {1: {'name': 'John', 'age': '27', 'sex': 'Male'},
          2: {'name': 'Marie', 'age': '22', 'sex': 'Female'},
          3: {'name': 'Luna', 'age': '24', 'sex': 'Female', 'married': 'No'}}
````

````python
people[4] = {'name': 'Peter', 'age': '29', 'sex': 'Male', 'married': 'Yes'}
print(people[4])
````

## Zip function

- Geographic Dataset - Latitude , Longitude , Curve - Polygon on the map
- It takes individual elements and combines them into pairs of elements

````python
# Zips multiple compound data type values together
````

````python
lat_ = [13.24, 78.912, 101.435]
lon_ = [12.81, 39.12, 58.72, 71]
population = [100, 500, 900]
# Index based zipping - default position based
````

````python
# (lat_, lon_) -> Geographical Point
# Location
````

````python
points = list(zip(lat_ , lon_, population))
````

# Sets

- Set is an unordered collection of data type that is iterable, mutable and has no duplicate elements. The order of elements in a set is undefined though it may consist of various elements.

- The major advantage of using a set, as opposed to a list, is that it has a highly optimized method for checking whether a specific element is contained in the set.

## Create

```python
# {key : value} - dictionary , {1,2,3} - set
s1 = {1,2,3,4,5}  # default it becomes a dictionary { : } - dictionary , { , } - set
s2 = set([1,2,3,4]) # explicit call

print(type(s1))
print(type(s2))
```

```python
s1
```

```python
s1 = {1,1,1,1,1,1,2}
```

```python
s1
```

```python
s1 # Unordered collection of Unique elements
```

```python
s2
```

- A set contains only unique elements but at the time of set creation, multiple duplicate values can also be passed. Order of elements in a set is undefined and is unchangeable. Type of elements in a set need not be the same, various mixed up data type values can also be passed to the set.

## Operators and Methods

- Most, though not quite all, set operations in Python can be performed in two different ways: by operator or by method. Let’s take a look at how these operators and methods work, using set union as an example.

* Once a set is created, you cannot change its items, but you can add new items.
- add() -- To add items from another set into the current set.

```python
# Sets are mutable -> Update a set
```

```python
s1 = {1,2,3,4,5}
s1
```

```python
s1.add(6)
s1
```

- remove() -- To remove a element from the Set by using it. But a KeyError arises if element doesn’t exist in the set. To remove elements from a set without KeyError, use discard(), if the element doesn’t exist in the set, it remains unchanged.

```python
s1.remove(10) # throws error when element not found
s1
```

```python
s1.discard(10) # Discards does not throw error if element does not exist
s1
```

- pop() -- remove the element from start of the set.

- issubset() -- In set theory, a set s1 is considered a subset of another set s2 if every element of s1 is in s2.

```python
# relative
# A set is a subset relative to another set
# When all values of s1 are also present in s2
# s2 is a subset of s1
# s1 is a superset of s2 -> all elements of s1 are present in s2 -
```

```python
s1 = {1,2,3,4,5}
s2 = {1,2,3}
```

```python
s2.issubset(s1)
```

```python
s1.issuperset(s2)
```

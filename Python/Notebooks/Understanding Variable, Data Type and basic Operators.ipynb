Certain operations will be performed before others in Python. This is similar to how operations such as addition, multipication are carried out in mathematics. This is known as orders of precedence. You may know the term BIDMAS or BODMAS to remember the order.

The following table shows the orders of precedence. Operators at the top will be carried out before those underneath.
```
Operators               Description
(, )                    Brackets / Parentheses
**                      Exponentiation / Power
*, /, %, //             Multiplication, division, modulus, floor division
+, -                    Addition, subtraction
<, <=, >, >=, !=, ==    Comparison operators
```

A patient is receiving a medication through a tube in their vein. The medication is in the form of a liquid solution, and for every 100 milliliters of the solution, there are 0.5 grams (or 500 milligrams) of the active medication. The medication is being given to the patient continuously for 8 hours, with the liquid dripping into their vein at a rate of 60 milliliters per hour. How many milligrams of the medication will the patient receive during these 8 hours?

````python
# Given values
infusion_rate = 60  # mL/h
solution_concentration = 0.5  # 0.5% concentration

# Convert solution concentration to mg/mL
medication_concentration = solution_concentration * 10  # 0.5% * 10 = 5 mg/mL

# Calculate the volume of the solution infused in 8 hours
time = 8  # hours
volume = infusion_rate * time  # mL

# Calculate the total amount of medication received
total_medication = volume * medication_concentration  # mg

# Print the result
print("The patient will receive", total_medication, "milligrams of the medication in 8 hours.")
````

### Multiplication

The * operator in Python can be utilized distinctly in the paired structure, which implies increase, restoring an outcome that is the standard arithmetic product result of its operands.

Example:
* (int, int) -> int 	4* 5 returns the result 20
* (float, float) -> float 	4.0 * 5.0 returns the result 20.0
* (complex, complex) -> complex 	4j * 5j returns the result (-20+0j)

````python
print(3*5)
print(3.0*5.0)
print(3j*5j)
````

### Division

The / Operator(cut) in Python can be utilized distinctly in the parallel structure, which implies division, restoring an outcome that is the standard number juggling the remainder of its operands: left operand partitioned by the right operand.

Example:
* / (int, int) -> int 	4 / 5 returns the result 0.8
* / (float, float) -> float 	4.0 / 5.0 returns the result 0.8
* / (complex, complex) -> complex 	4j / 5j returns the result (.08+0j)

````python
print(3/5)
print(3.0/5.0)
print(3j/5j)
````

### Floor Division

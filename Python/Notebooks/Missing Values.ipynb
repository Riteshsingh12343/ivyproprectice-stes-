# Missing Values
A missing value refers to the absence of a data entry in a dataset where a value is expected. This can occur due to various reasons, such as data entry errors, incomplete data collection, or intentional omissions. Missing values can impact data analysis and machine learning models, requiring strategies like imputation, deletion, or interpolation to handle them effectively.

* Missing values might be the most undesired values in data science. We definitely do not want to have them. However, they are always around.
* Since it is not reasonable to ignore missing values, we need to find ways to handle them efficiently and properly.
* In this Notebook, we will learn different methods to make the missing values go away without causing a lot of trouble. However, which method fits best to a particular situation depends on the data and task.

## Detecting Missing Values

Before handling missing values, we need to detect them. We use Pandas to check for missing values in a dataset.

````python
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame({
    "Date": pd.date_range(start="2022-01-01",periods=10,freq="D"),
    "Item": np.random.randint(1010,1020,size=10),
    "Measure_1": np.random.randint(1,10,size=10),
    "Measure_2": np.random.random(10).round(2),
    "Measure_3": np.random.random(10).round(2),
    "Measure_4": np.random.randn(10)
})
````

````python
df
````

The next step is to add the missing values. We will use the loc method to select the row and column combinations and make them equal to “np.nan” which is one of the standard missing value representations.

````python
df.loc[[2,7],"Item"] = np.nan
df.loc[[1,6,7],"Measure_1"] = np.nan
df.loc[[4,5,7],"Measure_2"] = np.nan
df.loc[[7],"Measure_3"] = np.nan
df.loc[[7,9],"Measure_4"] = np.nan
````

````python
df
````

- Ok in this case the data is very low so we can see it in naked eye what if there is too many data how can we detect it

````python
# Checking for missing values
print("\nMissing Values Count:\n", df.isnull().sum())  # Count of missing values per column
````

## Drop rows or columns that have a missing value
* One option is to drop the rows or columns that contain a missing value.

````python
df
````

* Only the date column does not have any missing values.

````python
df.dropna()
````

* With the default parameter values, the dropna function drops the rows that contain any missing value.
* There is only three row in the data frame that does not have any missing values.
* We can also choose to drop columns that have at least one missing value by using the axis parameter.

## Drop rows or columns that only have missing values
Another situation is to have a column or row that is full of missing values. Such columns or rows are useless so we can drop them.
The dropna function can be used for this as well. We just need to change the value of how parameter.

````python
# Add a column with missing value
df['null_column'] = np.nan
df
````

````python
df.dropna(how="all")
````

* Since the data frame does not have a row full of missing values, no row has been dropped.

## Drop rows or columns based on a threshold value
Dropping based on “any” or “all” is not always the best option. We sometimes need to drop rows or columns with “lots of” or “some” missing values.
We cannot assign such expressions to the how parameter but Pandas gives us a more accurate way which is the thresh parameter.
For instance, “thresh=4” means that the rows that have at least 4 non-missing values will be kept. The other ones will be dropped.
Our data frame has 6 columns so the rows that have 3 or more missing values will be dropped.

````python
df.dropna(thresh=4)
````

Only the seventh row had more than 4 missing values so it was the only one dropped.

## Drop based on a particular subset of columns
We can take only some of the columns into consideration when dropping columns.
The subset parameter of the dropna function is used for this task. For instance, we can drop the rows that have a missing value in measure_1 or measure_2 columns as follows:

````python
df.dropna(subset=["Measure_1","Measure_2"])
````

* Dropping is not the only option. In some cases, we may choose to fill missing values instead of dropping them.
* In fact, the filling might be a better option since data means value. How to fill the missing values, of course, depend on the structure of data and the task.

## Treat Missing Values

So far we have taken the critical step of dropping rows with missing values, ensuring that our dataset remains complete. This was a straightforward approach that helped us eliminate the immediate concern of incomplete data, but now, we face the next challenge: how can we handle the missing values in a more thoughtful way?

Dropping missing values might not always be the best solution. While it works in cases where the missing data is minimal, we lose valuable information when we discard entire rows. In fact, if a dataset has a significant number of missing values, dropping those rows could lead to biases or distort the underlying trends in the data.

To tackle this, we need a more robust way of handling missing data. Instead of just dropping rows, we can treat the missing values by filling them in with more informed methods. Let’s explore some options.

## Fill with a constant value
We can choose a constant value to be used as a replacement for the missing values.
If we just give one constant value to the fillna function, it will replace all the missing values in the data frame with that value.
A more reasonable method is to determine separate constant values for different columns. We can write them in a dictionary and pass it to the values parameter.

````python
df.fillna(value=0)
````

- But this is wrong since each column in the dataset may contain different types of information, whether numerical, categorical, or even time-based.

````python
values = {"Item":1016, "Measure_1":0}

df.fillna(value = values)
````

* The missing values in the item column are replaced with 1016 and the ones in the measure_1 column are replaced with 0.

- Again there is a problem how should you kown which value is correct to treat it, As a result, the way we handle missing values should be tailored to each specific column.

**For numerical columns, we might:**
- Use mean imputation to fill in missing values, but only if the data is approximately normally distributed.
- Use median imputation when the data is skewed or contains outliers, to avoid distorting the overall distribution.

**For categorical columns, we can:**
- Use mode imputation (the most frequent value) to fill in missing categories, ensuring that the general trend of the dataset is preserved.

## Fill with an aggregated value
Another option is to use an aggregated value such as mean, median, or mode.
The following line of code replaces the missing values in the measure 2 column with the average value of this column.

````python
df["Measure_2"].fillna(df["Measure_2"].mean())
````

## Replace with the previous or next value
It is possible to replace the missing values in a column with the previous or next value in that column.
This method might come in handy when working with time-series data. Consider you have a data frame that contains the daily temperature measurement and the temperate in one day is missing. The optimal solution would be to use the temperature in the next or previous day.
The method parameter of the fillna function is used for performing this task.

````python
df.fillna(method='ffill')
````

* The “bfill” fills the missing values backward so they are replaced with the next value.
Take a look at the last column. The missing values are replaced up to the first row. This may not be suitable for some cases.
Thankfully, we can limit the number of missing values replaced with this method. If we set the limit parameter as 1, then a missing value can only be replaced with its next value. The second or third following value will not be used for replacement.

````python
df.fillna(method='bfill', limit=1)
````

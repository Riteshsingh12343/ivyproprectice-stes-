# Read the data from the Table

```python
import cx_Oracle

# Replace with your Oracle database credentials
dsn_tns = cx_Oracle.makedsn('hostname', 'port', service_name='servicename')
conn = cx_Oracle.connect(user='yourusername', password='yourpassword', dsn=dsn_tns)

print("Successfully connected to Oracle Database")
```

## Creating a Table

```python
create_table_query = '''
CREATE TABLE transactions (
    transaction_id NUMBER GENERATED BY DEFAULT AS IDENTITY,
    transaction_date DATE,
    amount NUMBER,
    description VARCHAR2(255),
    PRIMARY KEY (transaction_id)
)
'''

cursor.execute(create_table_query)
conn.commit()
print("Table 'transactions' created")
```

**Insert a Single Record**<br>
In this case, we'll omit the transaction_id from the insert query, allowing the database to generate it automatically.

```python
insert_query_single = '''
INSERT INTO transactions (transaction_date, amount, description)
VALUES (:1, :2, :3)
'''

single_record = ('2024-06-08', 100.50, 'Grocery shopping')

cursor.execute(insert_query_single, single_record)
conn.commit()
print("Single record inserted into 'transactions'")
```

**Insert Multiple Rows**

```python
# Define the insert query
insert_query = '''
INSERT INTO transactions (transaction_date, amount, description)
VALUES (:1, :2, :3)
'''

# Sample data to insert (list of tuples)
data_to_insert = [
    ('2024-06-08', 100.50, 'Grocery shopping'),
    ('2024-06-09', 200.75, 'Electronics purchase'),
    ('2024-06-10', 50.25, 'Books purchase')
]

# Execute the insert query for multiple rows
cursor.executemany(insert_query, data_to_insert)

# Commit the transaction
conn.commit()

print("Multiple records inserted into 'transactions'")
```

**Insert Multiple Records from CSV File**<br>
Similarly, when inserting multiple records from a CSV file, we'll omit the transaction_id column from the insert query.

```python
import pandas as pd

# Define the insert query
insert_query = '''
INSERT INTO transactions (transaction_date, amount, description)
VALUES (:1, :2, :3)
'''

# Path to the CSV file
csv_file_path = 'transactions.csv'

# Read data from the CSV file using Pandas
data = pd.read_csv(csv_file_path)

# Iterate over rows and insert data into the database
for index, row in data.iterrows():
    cursor.execute(insert_query, (row['transaction_date'], row['amount'], row['description']))

# Commit the transaction
conn.commit()

print("Multiple records inserted from CSV file into 'transactions'")
```

```python
# Close the cursor and connection
cursor.close()
conn.close()
```

## Creating a table

```python
# Define the create table query for medical records
create_table_query = '''
CREATE TABLE medical_records (
    patient_id INT AUTO_INCREMENT PRIMARY KEY,
    patient_name VARCHAR(255),
    diagnosis VARCHAR(255),
    treatment VARCHAR(255)
)
'''

# Execute the create table query
cursor.execute(create_table_query)
print("Table 'medical_records' created")

# Commit the transaction
conn.commit()
```

## Inserting Data into a Table
**Insert a Single Row**

```python
# Define the insert query for a single record
insert_query_single = '''
INSERT INTO medical_records (patient_name, diagnosis, treatment)
VALUES (%s, %s, %s)
'''

# Insert a single record of data
single_record = ('John Doe', 'Fever', 'Prescription medication')
cursor.execute(insert_query_single, single_record)
print("Single record inserted into 'medical_records'")

# Commit the transaction
conn.commit()
```

**Insert Multiple Rows**

```python
# Define the insert query for multiple records
insert_query_multiple = '''
INSERT INTO medical_records (patient_name, diagnosis, treatment)
VALUES (%s, %s, %s)
'''

# Sample data to insert (list of tuples)
data_to_insert = [
    ('Jane Smith', 'Common cold', 'Rest and fluids'),
    ('Alice Johnson', 'Fractured arm', 'Surgery and cast')
]

# Execute the insert query for multiple records
cursor.executemany(insert_query_multiple, data_to_insert)
print("Multiple records inserted into 'medical_records'")

# Commit the transaction
conn.commit()
```

**Insert Multiple Records from CSV File**<br>
Similarly, when inserting multiple records from a CSV file, we'll omit the transaction_id column from the insert query.

```python
# Path to the CSV file
csv_file_path = 'medical_records.csv'

# Read data from the CSV file using Pandas
data_from_csv = pd.read_csv(csv_file_path)

# Iterate over rows and insert data into the database
for index, row in data_from_csv.iterrows():
    cursor.execute(insert_query_single, (row['patient_name'], row['diagnosis'], row['treatment']))

# Commit the transaction
conn.commit()
print("Multiple records inserted from CSV file into 'medical_records'")

# Close the cursor and connection
cursor.close()
conn.close()
```

## Read the data from table

```python
import pyodbc

# Replace with your SQL Server database credentials
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=yourserver;'
    'DATABASE=yourdatabase;'
    'UID=yourusername;'
    'PWD=yourpassword'
)

print("Successfully connected to SQL Server Database")
```

```python
cursor = conn.cursor()
cursor.execute("SELECT * FROM your_table")

for row in cursor:
    print(row)

cursor.close()
conn.close()
```

```python
# Define the create table query for financial transactions
create_table_query = '''
CREATE TABLE FinancialTransactions (
    TransactionID INT PRIMARY KEY,
    AccountID INT,
    TransactionDate DATE,
    Amount DECIMAL(10, 2)
)
'''

# Execute the create table query
cursor.execute(create_table_query)
print("Table 'FinancialTransactions' created")

# Commit the transaction
conn.commit()
```

## Inserting Data into a Table
**Insert a Single Row**

```python
# Define the insert query for a single transaction
insert_query_single = '''
INSERT INTO FinancialTransactions (TransactionID, AccountID, TransactionDate, Amount)
VALUES (?, ?, ?, ?)
'''

# Insert a single transaction record
single_transaction_record = (1, 101, '2024-06-08', 500.00)
cursor.execute(insert_query_single, single_transaction_record)
print("Single transaction record inserted into 'FinancialTransactions'")

# Commit the transaction
conn.commit()
```

**Insert Multiple Rows**

```python
# Define the insert query for multiple transactions
insert_query_multiple = '''
INSERT INTO FinancialTransactions (TransactionID, AccountID, TransactionDate, Amount)
VALUES (?, ?, ?, ?)
'''

# Sample data to insert (list of tuples)
data_to_insert = [
    (2, 102, '2024-06-09', 750.25),
    (3, 101, '2024-06-10', 1000.50)
]

# Execute the insert query for multiple transactions
cursor.executemany(insert_query_multiple, data_to_insert)
print("Multiple transaction records inserted into 'FinancialTransactions'")

# Commit the transaction
conn.commit()
```

```python
# Path to the CSV file containing transaction data
csv_file_path = 'financial_transactions.csv'

# Read data from the CSV file using Pandas
data_from_csv = pd.read_csv(csv_file_path)

# Iterate over rows and insert data into the database
for index, row in data_from_csv.iterrows():
    cursor.execute(insert_query_single, (row['TransactionID'], row['AccountID'], row['TransactionDate'], row['Amount']))

# Commit the transaction
conn.commit()
print("Multiple transaction records inserted from CSV file into 'FinancialTransactions'")

# Close the cursor and connection
cursor.close()
conn.close()
```
